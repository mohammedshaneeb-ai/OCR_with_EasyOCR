## In this Notebook I am showing you How you can extract text from an image using EasyOCR

```python
import easyocr
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Reading Image
IMAGE_PATH = 'stop.jpg'
```

```python
# Creating instance
reader = easyocr.Reader(['en'],gpu=False)
```

```python
# Reading Texts from image
result = reader.readtext(IMAGE_PATH)
```

```python
img = cv2.imread(IMAGE_PATH)
plt.imshow(img)
plt.show()
```

```python
result
```

```python
top_left = list(result[0][0][0])
bottom_right = list(result[0][0][2])
text = result[0][1]
font = cv2.FONT_HERSHEY_PLAIN
```

```python
img = cv2.imread('stop.jpg')
img = cv2.rectangle(img,top_left,bottom_right,(0,255,0),3)
img = cv2.putText(img,text,top_left, font, 5,(255,0,0),5,cv2.LINE_AA)
plt.imshow(img)
plt.show()
```

# Reading Multiple lines

```python
IMAGE_PATH = ('poem1.jpeg')
```

```python
reader = easyocr.Reader(['en'],gpu=False)
result = reader.readtext(IMAGE_PATH)
```

```python
img = cv2.imread(IMAGE_PATH)
spacer = 100
doc = []
for detection in result: 
    top_left = list(detection[0][0])
    bottom_right = list(detection[0][2])
    
    top_left = [int(i) for i in top_left]
    bottom_right = [int(i) for i in bottom_right]
    print(top_left)
    print(bottom_right)
    print('working')
    text = detection[1]
    doc.append(text)
    print(text)
    img = cv2.rectangle(img,top_left,bottom_right,(0,255,0),3)
    img = cv2.putText(img,text,(20,spacer), font, 3,(0,0,255),2,cv2.LINE_AA)
    spacer+=15
    
plt.imshow(img)
plt.show()
```
